```python
import pandas as pd
import os

data = pd.read_csv(os.path.join(os.getcwd(), 'training_data', '1_output_training_data.csv')).to_dict(orient='records')

training_dict = {
    'epoch': [], 
    'loss': [], 
    'train_acc': [], 
    'val_acc': [],
    'train_f1': [],
    'val_f1': []
}

for i in range(len(data)):
    training_dict['epoch'].append(data[i]['epoch'])
    training_dict['loss'].append(data[i]['loss'])
    training_dict['train_acc'].append(data[i]['train_acc'])
    training_dict['val_acc'].append(data[i]['val_acc'])
    training_dict['train_f1'].append(data[i]['train_f1'])
    training_dict['val_f1'].append(data[i]['val_f1'])
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# Plot the Loss
axs[0].plot(training_dict['epoch'], training_dict['loss'], label='Loss')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Loss')
axs[0].legend()

# Plot the metrics
axs[1].plot(training_dict['epoch'], training_dict['train_acc'], label='Train Accuracy')
axs[1].plot(training_dict['epoch'], training_dict['val_acc'], label='Validation Accuracy')
axs[1].plot(training_dict['epoch'], training_dict['train_f1'], label='Train F1 Score')
axs[1].plot(training_dict['epoch'], training_dict['val_f1'], label='Validation F1 Score')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Metrics')
axs[1].legend()

plt.show()
```
